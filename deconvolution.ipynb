```python
import os
import numpy as np
import scipy as sp
import scipy.stats as stats
import scipy.signal as signal
import matplotlib.pyplot as plt
from sympy import *
import math
```

```python
%matplotlib inline 
```

```python
# define an Impulse Response Function:
def pupil_IRF(t=0, s=1.0/(10**26), n=10.1, tmax=0.93, IRF_len=3):
    
    """ pupil_IRF defines the IRF (response to a transient input) of the pupil.
    
    Parameters
    ----------
    t: IRF is defined with respect to time 't'
    s: scaling factor
    n: sets the width
    tmax: sets the time to peak 
    IRF_len: function is evaluated for t = [0:IRF_len]
    
    Returns
    -------
    y: IRF, evaluated for t = [0:IRF_len]
    yprime: IRF first derivative, evaluated for t = [0:IRF_len]
    
    """
    
    # in sympy:
    t = Symbol('t')
    y = ( (s) * (t**n) * (math.e**((-n*t)/tmax)) )
    yprime = y.diff(t)
    
    # lambdify:
    y = lambdify(t, y, "numpy")
    yprime = lambdify(t, yprime, "numpy")
    
    # evaluate:
    y = y(np.linspace(0,IRF_len,IRF_len*10))
    yprime = yprime(np.arange(IRF_len))
    
    return (y, yprime)

# create the IRF:
IRF_len = 3
IRF, IRF_prime = pupil_IRF(IRF_len=IRF_len)
IRF = IRF / IRF.max()

# plot the IRF:
fig = plt.figure()
plt.plot(np.linspace(0,IRF_len,len(IRF)), IRF, color='r')
plt.legend(['IRF'])
plt.title('Impulse Response Function')
plt.xlabel('time (s)')
plt.ylabel('a.u.')
```

```python
# input:
nr_timepoints = 350
times = [10,60,110,115,160,165,210,220,260,270]
input_signal = np.zeros(nr_timepoints)
for i in times:
    input_signal[i] = 1

# plot input:
time = np.linspace(0,nr_timepoints/10.0,nr_timepoints)
fig = plt.figure()
plt.plot(time, input_signal, 'r')
plt.legend(['input signal'], loc=2)
plt.title('input signal')
plt.xlabel('time (s)')
plt.ylabel('a.u.')
    
# convolve inputs with IRF:    
convolved_signal = (sp.convolve(input_signal, IRF, 'full'))[:-(IRF.shape[0]-1)]

# plot simulated convolved signal without noise:
fig = plt.figure()
plt.plot(time, input_signal, 'r')
plt.plot(time, convolved_signal)
plt.legend(['input signal', 'simulated convolved signal'], loc=2)
plt.title('simulated convolved signal without noise')
plt.xlabel('time (s)')
plt.ylabel('a.u.')

# let's add some noise:
convolved_signal = convolved_signal + np.random.normal(0,0.25,len(convolved_signal))

# plot simulated convolved signal with noise:
fig = plt.figure()
plt.plot(time, input_signal, 'r')
plt.plot(time, convolved_signal)
plt.legend(['input signal', 'simulated convolved signal'], loc=2)
plt.title('simulated convolved signal with noise')
plt.xlabel('time (s)')
plt.ylabel('a.u.')
```

```python
# mean response:
epochs = np.vstack([convolved_signal[i:i+30] for i in times])
mean_response = np.mean(epochs, axis=0)

# plot mean response versus IRF:
time = np.linspace(0,len(IRF)/10.0,len(IRF))
fig = plt.figure()
plt.plot(time, mean_response, color='b')
plt.plot(time, IRF, color='r')
plt.legend(['mean response', 'true response'])
plt.title('mean response')
plt.xlabel('time')
plt.ylabel('a.u.')
```

```python
# make design matrix:
nr_samples = 30 # here we define the length of the deconvolution response we're interested in (30 samples = 3-s in our case).
designMatrix = np.zeros((nr_samples,nr_timepoints))
for i in range(len(times)):
    for j in range(nr_samples):
        designMatrix[j,times[i]+j] = 1

# plot design matrix:
fig = plt.figure()
plt.imshow(designMatrix.T, cmap='gray')
plt.xticks([0,35])
plt.title('design matrix')
plt.xlabel('nr samples')
plt.ylabel('length run')
```

```python
# deconvolution:
designMatrix = np.mat(designMatrix).T
deconvolved_response = ((designMatrix.T * designMatrix).I * designMatrix.T) * np.mat(convolved_signal).T
deconvolved_response = np.array(deconvolved_response)

# plot deconvoled response versus true response:
fig = plt.figure()
fig.add_subplot(211)
plt.plot(np.linspace(0,IRF_len,len(IRF)),deconvolved_response, color='b')
plt.xlim(xmax=3)
plt.ylim(ymin=0)
plt.legend(['deconvolved response'])
plt.title('deconvolved response')
plt.xlabel('time (s)')
plt.ylabel('a.u.')
fig.add_subplot(212)
plt.plot(np.linspace(0,IRF_len,len(IRF)),IRF, color='r')
plt.xlim(xmax=3)
plt.ylim(ymin=0)
plt.legend(['true response'])
plt.xlabel('time (s)')
plt.ylabel('a.u.')
```
